Task - 3 - Outlier Detection:
Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 × IQR rule.
3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.
4. Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'South_Asia_Dataset.csv'
df = pd.read_csv(file_path)


# Task 3: Outlier Detection using 1.5 * IQR Rule

# Calculate the IQR for Score and Log GDP per Capita
score_q1 = south_asia_df['score'].quantile(0.25)
score_q3 = south_asia_df['score'].quantile(0.75)
score_iqr = score_q3 - score_q1

gdp_q1 = south_asia_df['Log GDP per capita'].quantile(0.25)
gdp_q3 = south_asia_df['Log GDP per capita'].quantile(0.75)
gdp_iqr = gdp_q3 - gdp_q1

# Define outlier thresholds
score_lower_bound = score_q1 - 1.5 * score_iqr
score_upper_bound = score_q3 + 1.5 * score_iqr

gdp_lower_bound = gdp_q1 - 1.5 * gdp_iqr
gdp_upper_bound = gdp_q3 + 1.5 * gdp_iqr

# Identify outliers
outliers = south_asia_df[
    (south_asia_df['score'] < score_lower_bound) | (south_asia_df['score'] > score_upper_bound) |
    (south_asia_df['Log GDP per capita'] < gdp_lower_bound) | (south_asia_df['Log GDP per capita'] > gdp_upper_bound)
]

# Display outliers
print("Outliers in South Asia based on Score and GDP per Capita:")
print(outliers)

# Visualize outliers using a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(south_asia_df['Log GDP per capita'], south_asia_df['score'], color='skyblue', label='Normal')
plt.scatter(outliers['Log GDP per capita'], outliers['score'], color='red', label='Outliers')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.title('Outlier Detection in South Asia (GDP vs Score)')
plt.legend()
plt.grid(True)
plt.show()
```

6. Outlier Detection:

• Identify outlier countries in both regions based on Score and GDP per Capita.

• Plot these outliers and discuss their implications.


```python
# Task 6: Outlier Detection

south_asia_df = pd.read_csv('South_Asia_Dataset.csv')
middle_east_df = pd.read_csv('Middle_East_Dataset.csv')


# Function to detect outliers using the 1.5 * IQR rule
def detect_outliers(df, metric):
    q1 = df[metric].quantile(0.25)
    q3 = df[metric].quantile(0.75)
    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    outliers = df[(df[metric] < lower_bound) | (df[metric] > upper_bound)]
    return outliers

# Detect outliers in both datasets
south_asia_outliers = detect_outliers(south_asia_df, 'score')
middle_east_outliers = detect_outliers(middle_east_df, 'score')

# Print outliers
print("South Asia Outliers based on Score:")
print(south_asia_outliers[['Country name', 'score', 'Log GDP per capita']])

print("\nMiddle East Outliers based on Score:")
print(middle_east_outliers[['Country name', 'score', 'Log GDP per capita']])

# Visualize the outliers using scatter plots
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 8))

# South Asia Scatter Plot
plt.subplot(2, 1, 1)
plt.scatter(south_asia_df['Log GDP per capita'], south_asia_df['score'], color='skyblue', label='Normal')
plt.scatter(south_asia_outliers['Log GDP per capita'], south_asia_outliers['score'], color='red', label='Outliers')
plt.title('South Asia: GDP vs Score (Outliers Highlighted)')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.legend()
plt.grid(True)

# Middle East Scatter Plot
plt.subplot(2, 1, 2)
plt.scatter(middle_east_df['Log GDP per capita'], middle_east_df['score'], color='lightgreen', label='Normal')
plt.scatter(middle_east_outliers['Log GDP per capita'], middle_east_outliers['score'], color='red', label='Outliers')
plt.title('Middle East: GDP vs Score (Outliers Highlighted)')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
